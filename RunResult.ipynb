```python

columns_name1=['System_info','Username','Admin_name','Admin_Lname','Admin_id',\
                            'name1','Lname1','id1',\
                            'name2','Lname2','id2',\
                           'name3','Lname3','id3',\
                           'name4','Lname4','id4',\
                           'name5','Lname5','id5',\
                           'name6','Lname6','id6',\
                           'name7','Lname7','id7']

columns_name2 = ['System_info','Username','name','Lname','id']

arr = os.listdir('C:\\git_project\\CU320\\UserReport\\Result')
arr_list = []
df_list = []

df_all = pd.DataFrame(columns=columns_name2)
```

```python

for i in arr:
    path = "C:\\git_project\\CU320\\UserReport\\Result\\" + i
    df_raw = pd.read_excel(path)
    df_raw = df_raw.iloc[2: , 1:]

    df4 = df_raw.set_axis(columns_name1, axis=1)
    
    df_guser = pd.DataFrame()

    df_admin = df4[['System_info','Username','Admin_name','Admin_Lname','Admin_id']]
    df_admin = df_admin.set_axis(columns_name2, axis=1)
    df_admin['Role'] = 'Admin'

    #df_guser = df_guser.append(df_admin, ignore_index = True)
    df_guser  = pd.concat([df_guser,df_admin],ignore_index=True)

    df_User1 = df4[['System_info','Username','name1','Lname1','id1']]
    df_User1 = df_User1.set_axis(columns_name2, axis=1)
    df_User1['Role'] = 'User1'

    #df_guser = df_guser.append(df_User1, ignore_index = True)
    df_guser  = pd.concat([df_guser,df_User1],ignore_index=True)

    df_User2 = df4[['System_info','Username','name2','Lname2','id2']]
    df_User2 = df_User2.set_axis(columns_name2, axis=1)
    df_User2['Role'] = 'User2'

    df_guser  = pd.concat([df_guser,df_User2],ignore_index=True)

    df_User3 = df4[['System_info','Username','name3','Lname3','id3']]
    df_User3 = df_User3.set_axis(columns_name2, axis=1)
    df_User3['Role'] = 'User3'

    df_guser  = pd.concat([df_guser,df_User3],ignore_index=True)

    df_User4 = df4[['System_info','Username','name4','Lname4','id4']]
    df_User4 = df_User4.set_axis(columns_name2, axis=1)
    df_User4['Role'] = 'User4'

    df_guser  = pd.concat([df_guser,df_User4],ignore_index=True)

    df_User5 = df4[['System_info','Username','name5','Lname5','id5']]
    df_User5 = df_User5.set_axis(columns_name2, axis=1)
    df_User5['Role'] = 'User5'

    df_guser  = pd.concat([df_guser,df_User5],ignore_index=True)

    df_User6 = df4[['System_info','Username','name6','Lname6','id6']]
    df_User6 = df_User6.set_axis(columns_name2, axis=1)
    df_User6['Role'] = 'User6'

    df_guser  = pd.concat([df_guser,df_User6],ignore_index=True)

    df_User7 = df4[['System_info','Username','name7','Lname7','id7']]
    df_User7 = df_User7.set_axis(columns_name2, axis=1)
    df_User7['Role'] = 'User7'

    df_guser  = pd.concat([df_guser,df_User7],ignore_index=True)
    
    #filter
    df_guser = df_guser[df_guser['name'].notnull()]
    df_guser = df_guser[~((df_guser['name'] == '----') | (df_guser['name'] == ' ') | (df_guser['name'] == ' '))]
    df_guser = df_guser.sort_values(by='Username')
    df_guser = df_guser.reset_index(drop=True)
    
    df_all = pd.concat([df_all,df_guser], axis=0, ignore_index=True)
          
    #exec(f'df_{guser} = pd.DataFrame()')
    #exec(f'df_{guser} = df_guser')
    
    #arr_list.append(i.split('.')[0])
    
    guser = i.split('.')[0]
    output_path = 'C:\\git_project\\CU320\\UserReport\\Resultsummary\\'
    filename = 'result_' + guser + '.xlsx'
        
    writer = pd.ExcelWriter(output_path+filename, engine='xlsxwriter')
    df_guser.to_excel(writer, sheet_name='Sheet1', index_label="No.")
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']
    writer.close()


    
output_path = 'C:\\git_project\\CU320\\UserReport\\Resultsummary\\'
filename = 'result_all' + '.xlsx'
#filter
df_all = df_all[~((df_all['name'] == '----') | (df_all['name'] == ' ') | (df_all['name'] == ' '))]
df_all = df_all.sort_values(by='Username')
df_all = df_all.reset_index(drop=True)

writer = pd.ExcelWriter(output_path+filename, engine='xlsxwriter')
df_all.to_excel(writer, sheet_name='Sheet1', index_label="No.")
workbook = writer.book
worksheet = writer.sheets['Sheet1']
writer.close()
```

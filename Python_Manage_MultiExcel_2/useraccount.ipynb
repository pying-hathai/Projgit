```python
focus_col = ['ระบบงาน','Username']

col1 = ['บุคลากรผู้ดูแล Username','ผู้ใช้งาน1','ผู้ใช้งาน2','ผู้ใช้งาน3','ผู้ใช้งาน4','ผู้ใช้งาน5','ผู้ใช้งาน6','ผู้ใช้งาน7']
multi_index = pd.MultiIndex.from_product([col1, ["ชื่อ","นามสกุล","รหัสพนักงาน"]])
df10 = pd.DataFrame(columns=multi_index)

for i in work_unit:
    df4 = df2[focus_col][df2['หน่วยงาน'] == i].sort_values(by=['ระบบงาน','Username'])
    data4 = {'ระบบงาน' : df4['ระบบงาน'].values, 'Username' : df4['Username'].values}
    
    df5 = pd.DataFrame(data4)
    # Create the new column level
    new_level = ['System_info']
    # Construct the new MultiIndex
    new_columns = pd.MultiIndex.from_product([new_level, df5.columns])
    # Assign the new MultiIndex
    df5.columns = new_columns
    df5.index += 1
    
    df510 = pd.concat([df5, df10], ignore_index=True)
    df510.index += 1
    
    
       
    filename = 'Dep_' + i + '.xlsx'
    
    writer = pd.ExcelWriter(filename, engine='xlsxwriter')
    df510.to_excel(writer, sheet_name='Sheet1', index_label="No.")
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']
    
    # set column width for 'Sheet1'. If need do the same for other Sheet 
    column_width = 15
    idx = 1
    worksheet.set_column(idx, 30, column_width)

    border_fmt = workbook.add_format({'bottom':2, 'top':2, 'left':2, 'right':2})
    worksheet.conditional_format(xlsxwriter.utility.xl_range(0, 0, 150, len(df510.columns)), {'type': 'no_errors', 'format': border_fmt})
    writer.save()

    #print(filename)
    #df510.to_excel(filename, index = True, index_label="No.")
    

```

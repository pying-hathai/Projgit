```python
cwd = os.getcwd()
path = os.path.join(cwd, "depinfo.xlsx")
path2 =  os.path.join(cwd, "memo.docx")
OUTPUT = os.path.join(cwd, "OUTPUT")

doc = Document(path2)
df = pd.read_excel(path, dtype={'รหัสหน่วยงาน': str})
df['CodeDoc2'] = df['CodeDoc'].apply(lambda i : i.split("/")[1])
df
```

```python
for i in list(range(len(df))):
#for i in [37,38]:
    
    print(i)
    
    name_Dep = df['Dep'][i]
    name_ShortLink = df['ShortLink'][i]
    name_Code = df['รหัสหน่วยงาน'][i]
    name_Email = df['Email'][i]
    name_CodeDoc = df['CodeDoc'][i]
    name_CodeDoc2 = df['CodeDoc2'][i]
    
    print(name_Dep)
    print(name_Code)
    
    new_data = []
    doc.tables[0].cell(0,1).paragraphs[0].text = name_CodeDoc
    doc.tables[0].cell(0,1).paragraphs[0].runs[0].font.name = 'TH Sarabun New'
    doc.tables[0].cell(0,1).paragraphs[0].runs[0].font.size = Pt(16)
    new_data.append(doc.tables[0].cell(0,1).paragraphs[0].text)
    
    doc.tables[1].cell(0,1).paragraphs[0].text = name_Dep
    doc.tables[1].cell(0,1).paragraphs[0].runs[0].font.name = 'TH Sarabun New'
    doc.tables[1].cell(0,1).paragraphs[0].runs[0].font.size = Pt(16)
    new_data.append(doc.tables[1].cell(0,1).paragraphs[0].text)
    
    doc.tables[2].cell(0,0).paragraphs[0].text = name_ShortLink
    doc.tables[2].cell(0,0).paragraphs[0].runs[0].font.name = 'TH Sarabun New'
    doc.tables[2].cell(0,0).paragraphs[0].runs[0].font.size = Pt(16)
    doc.tables[2].cell(0,0).paragraphs[0].runs[0].font.bold = True
    new_data.append(doc.tables[2].cell(0,0).paragraphs[0].text)

    file_name_docx =  str(name_CodeDoc2) + '.docx'
    doc.save(os.path.join(OUTPUT, file_name_docx))
    
    #####DOCtoPDF#####
    file_name_pdf =  str(name_CodeDoc2) + '.pdf'
    inputFile = cwd + "\\OUTPUT\\" + file_name_docx
    outputFile = cwd + "\\OUTPUT\\" + file_name_pdf
    convert(inputFile, outputFile)
```
